```python
df['winner'].value_counts().head(10)
```

```python
import matplotlib.pyplot as plt

df['winner'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 IPL Winning Teams')
plt.xlabel('Team')
plt.ylabel('Wins')
plt.xticks(rotation=45)
plt.show()
```

```python
df['player_of_match'].value_counts().head(10)
```

```python
df['player_of_match'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Player of the Match')
plt.xlabel('Player')
plt.ylabel('Awards')
plt.xticks(rotation=45)
plt.show()
```
